###### Question_4) How has the construction of roller coasters (speed, height, inversions) evolved over the years?

```python
data['Opening date'] = pd.to_datetime(data['Opening date'], errors='coerce')
data['Year'] = data['Opening date'].dt.year

time_series_trends = data.groupby('Year')[['speed_mph', 'height_ft', 'Inversions_clean']].mean()

time_series_trends.dropna(inplace=True) #dropping Nan Values

#The trends over time
time_series_trends.plot(figsize=(12, 6), title='Roller Coaster Construction Trends Over Time')
plt.show()
```
